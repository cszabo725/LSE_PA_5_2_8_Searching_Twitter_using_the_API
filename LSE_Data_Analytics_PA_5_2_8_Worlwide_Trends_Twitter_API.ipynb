```python
# Import libraries.
import yaml
from yaml.loader import SafeLoader

# Import the yaml file - remember to specify the whole path.
twitter_creds = yaml.safe_load(open('twitter_tmp.yaml', 'r').read())

# View the keys in the dictionary.
twitter_creds.keys()

# Import the library.
from twitter import *

# Pass your twitter credentials.
twitter_api = Twitter(auth=OAuth(twitter_creds['access_token'],
                                 twitter_creds['access_token_secret'], 
                                 twitter_creds['api_key'],
                                 twitter_creds['api_secret_key']))

# Confirm your connection.
print(twitter_api)
```

```python
# Create the variable
ww_trends = twitter_api.trends.available()

# How many trends are available
print(len(ww_trends))

# Example of ww_trends
ww_trends[0]
```

```python
# Identify a specific cities - London and New York
london = 'London'
new_york = 'New York'

# Create a variable
london_city = [_ for _ in ww_trends if _['name'] == london]
new_york_city = [_ for _ in ww_trends if _['name'] == new_york]

# View the ouput
print(len(london_city))
print(len(new_york_city))

print(london_city[0])
print(new_york_city[0])
```

```python
# Twitter trends in London
london_trends = twitter_api.trends.place(_id = london_city[0]['woeid'])

# View the output
london_trends
```

```python
# View output as a DataFrame
# Import Pandas
import pandas as pd

# Create a DataFrame
london_trends_pd = pd.DataFrame(london_trends[0]['trends'])

# View the DataFrame
london_trends_pd
```

```python
# Narrow list down to over 50,000 tweets
london_trends_over50k_pd = london_trends_pd[london_trends_pd['tweet_volume'] > 50000]\
.sort_values('tweet_volume', ascending=False)

# View the output
print(london_trends_over50k_pd.shape)
london_trends_over50k_pd
```

```python
# Twitter trends in New York City
new_york_trends = twitter_api.trends.place(_id = new_york_city[0]['woeid'])

# View the output
new_york_trends
```

```python
# Create a DataFrame
new_york_trends_pd = pd.DataFrame(new_york_trends[0]['trends'])

# View the DataFrame
new_york_trends_pd
```

```python
# Narrow list down to over 50,000 tweets
new_york_trends_over50k_pd = new_york_trends_pd[new_york_trends_pd['tweet_volume'] > 50000]\
.sort_values('tweet_volume', ascending=False)

# View the output
print(new_york_trends_over50k_pd.shape)
new_york_trends_over50k_pd
```

```python
# Find common topics
london_trends_list = [trend['name'] for trend in london_trends[0]['trends']]
new_york_trends_list = [trend['name'] for trend in new_york_trends[0]['trends']]

# View the output
print(london_trends_list)
print(new_york_trends_list)
```

```python
# Find common trends between cities
london_trends_set = set(london_trends_list)
new_york_trends_set = set(new_york_trends_list)

# Set the variable
common_trends = london_trends_set.intersection(new_york_trends_set)

# View the output
print(common_trends)
```
